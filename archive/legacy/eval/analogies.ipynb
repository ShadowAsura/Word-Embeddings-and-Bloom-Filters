```python
import json
import numpy as np
import os
import glob

# Load vectors: prefer window_4_iter_200 (from iterative_vectors.py); fall back to any available file
vector_dir = 'data/iterative_vectors'
if not os.path.isdir(vector_dir):
    vector_dir = '../data/iterative_vectors'
window, iteration = 4, 200
vector_file = os.path.join(vector_dir, f'window_{window}_iter_{iteration}.json')
if not os.path.isfile(vector_file):
    found = sorted(glob.glob(os.path.join(vector_dir, 'window_*_iter_*.json')),
                   key=lambda p: int(p.split('_iter_')[1].replace('.json', '')))
    if found:
        vector_file = found[-1]
        print(f"Using {vector_file}")
    else:
        raise FileNotFoundError(
            f"No iterative vector files in {os.path.abspath(vector_dir)}.\n"
            "Generate them by running from project root: python iterative_vectors.py"
        )
with open(vector_file) as f:
    vectors = json.load(f)
```

```python
import sys
# Ensure project root is on path so "from eval.evaluate" works when notebook is in eval/
_root = os.path.abspath(os.curdir)
if os.path.basename(_root) == 'eval':
    _root = os.path.dirname(_root)
if _root not in sys.path:
    sys.path.insert(0, _root)

from eval.evaluate import evaluate_vectors

# Build vocabulary and embedding matrix
vkeys = sorted(list(vectors.keys()))
vocab = {vkeys[i]: i for i in range(len(vkeys))}

# Convert to numpy array
W = np.array([vectors[word] for word in vkeys])

# Normalize vectors to unit length
W_norm = np.zeros(W.shape)
d = (np.sum(W ** 2, 1) ** (0.5))
W_norm = (W.T / d).T

# Run evaluation
evaluate_vectors(W=W_norm, vocab=vocab)
```

```python
vectors = np.array([[0,2,5,6,2.5],[0.5,83.2,6.7,4.3,4],[10,4,5,2,7]])
vectors = vectors / np.linalg.norm(vectors, axis=1, keepdims=True) # normalize along rows (words)
minimums = vectors.min(axis=0)
vectors = (vectors - minimums) / (vectors.max(axis=0) - minimums) # normalize along columns (dimensions)
```
