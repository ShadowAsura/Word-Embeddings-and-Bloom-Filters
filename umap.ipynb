## Unifold Manifold Approximation and Projection

Uses kqmw_iteration.json to generate plots

```python
import json
import numpy as np
import matplotlib.pyplot as plt
import umap

window = 4
fpath = f'data/kqmw_iteration_window_{window}.json'
word = 'king' 

with open(fpath, 'r') as f:
    kqmw_iteration = json.load(f)

vectors = np.array(kqmw_iteration[word])

reducer = umap.UMAP(n_components=2)
embedding = reducer.fit_transform(vectors)

fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(embedding[:, 0], embedding[:, 1], s=5)

ax.set_xlabel('UMAP 1')
ax.set_ylabel('UMAP 2')

title = f"UMAP of '{word}' Embeddings"
subtitle = f"Window = {window}"
ax.set_title(f"{title}\n{subtitle}", fontsize=12)
plt.savefig(f'plots/{word}_umap_window_{window}.png', dpi=300)
plt.show()
```

3D version

```python
window = 4
fpath = f'data/kqmw_iteration_window_{window}.json'
word = 'king'

with open(fpath, 'r') as f:
    kqmw_iteration = json.load(f)

vectors = np.array(kqmw_iteration[word])

reducer_3d = umap.UMAP(n_components=3)
embedding_3d = reducer_3d.fit_transform(vectors)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(embedding_3d[:, 0], embedding_3d[:, 1], embedding_3d[:, 2], s=5)
ax.set_xlabel('UMAP 1')
ax.set_ylabel('UMAP 2')
ax.set_zlabel('UMAP 3')


title = f"3D UMAP of '{word}' Embeddings"
subtitle = f"Window = {window}"

ax.set_title(f"{title}\n{subtitle}", fontsize=12)

plt.savefig(f'plots/{word}_umap3D_window_{window}.png', dpi=300)
plt.show()
```
