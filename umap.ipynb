## Unifold Manifold Approximation and Projection

Uses kqmw_iteration.json to generate plots

```python
import json
import numpy as np
import matplotlib.pyplot as plt
import umap


word = 'king' 

with open('data/kqmw_iteration.json', 'r') as f:
    kqmw_iteration = json.load(f)

vectors = np.array(kqmw_iteration[word])

reducer = umap.UMAP(n_components=2)
embedding = reducer.fit_transform(vectors)

plt.figure(figsize=(8, 6))
plt.scatter(embedding[:, 0], embedding[:, 1], s=5)
plt.title(f'2D UMAP - {word}')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()
```

3D version

```python
word = 'king'  

with open('data/kqmw_iteration.json', 'r') as f:
    kqmw_iteration = json.load(f)

vectors = np.array(kqmw_iteration[word])

from mpl_toolkits.mplot3d import Axes3D

reducer_3d = umap.UMAP(n_components=3)
embedding_3d = reducer_3d.fit_transform(vectors)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(embedding_3d[:, 0], embedding_3d[:, 1], embedding_3d[:, 2], s=5)
ax.set_title(f'3D UMAP - {word}')
ax.set_xlabel('UMAP 1')
ax.set_ylabel('UMAP 2')
ax.set_zlabel('UMAP 3')
plt.show()
```
